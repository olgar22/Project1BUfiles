```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%store -r one_mil_df
one_mil_df
```

```python
# calculate % change year-to-year for 1 mil and more cities
#col_names =  ['Zip', 'City','2011', '2012', '2013', '2014', '2015', '2016', '2017','All']
my_df=pd.DataFrame()
my_df['Zip']=one_mil_df['Short Zip']
my_df['City'] = one_mil_df['Metro Area']
my_df['2011']=(one_mil_df['Population 2011']- one_mil_df['Population 2010'])/one_mil_df['Population 2010'] *100
my_df['2012']=(one_mil_df['Population 2012']- one_mil_df['Population 2010'])/one_mil_df['Population 2010'] *100
my_df['2013']=(one_mil_df['Population 2013']- one_mil_df['Population 2010'])/one_mil_df['Population 2010'] *100
my_df['2014']=(one_mil_df['Population 2014']- one_mil_df['Population 2010'])/one_mil_df['Population 2010'] *100
my_df['2015']=(one_mil_df['Population 2015']- one_mil_df['Population 2010'])/one_mil_df['Population 2010'] *100
my_df['2016']=(one_mil_df['Population 2016']- one_mil_df['Population 2010'])/one_mil_df['Population 2010'] *100
my_df['2017']=(one_mil_df['Population 2017']- one_mil_df['Population 2010'])/one_mil_df['Population 2010'] *100
my_df['Total Growth']=(one_mil_df['Population 2017']- one_mil_df['Population 2010'])/one_mil_df['Population 2010'] *100
my_df['Population 2017']=one_mil_df['Population 2017']

#my_df
```

```python
my_df
```

```python
#we are only interested in the following: 10 highest jumps, 10 largest drops, and the US trend 
jump_df=my_df.nlargest(12, 'Total Growth')
drop_df = my_df.nsmallest(10, 'Total Growth')
cities_of_interest = pd.concat([jump_df, drop_df], ignore_index=True)
cities_of_interest_short =cities_of_interest[['Zip', 'City', 'Total Growth', 'Population 2017']]
cities_of_interest_short = cities_of_interest_short.drop([4 , 13])
cities_of_interest_short=cities_of_interest_short.reset_index(drop=True)
#cities_of_interest_short.sort_values(by=['Total Growth'])
cities_of_interest_short = cities_of_interest_short.sort_values(by=['Total Growth'], axis=0, ascending=False)
print(cities_of_interest_short)
len(cities_of_interest)
```

```python
# Export file as a CSV, with the header
try:
    cities_of_interest_short.to_csv("Output/cities_of_interest_short.csv", index=False, header=True)
except:
    print('Directory does not exist or file not accessible')
len(cities_of_interest_short)
```
