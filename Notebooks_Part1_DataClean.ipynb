```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Store filepath in a variable
file_one = "Resources/PEP_2017_GCTPEPANNR.US23PR.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
file_one_df = pd.read_csv(file_one, encoding="ISO-8859-1")
```

```python
# Show just the header
file_one_df.head()
```

```python
#drop extraneous columns
file_one_df.drop(file_one_df.columns[0:4],axis=1,inplace=True)
file_one_df.head()
```

```python
#drop extraneous columns
file_one_df.drop(file_one_df.columns[3:4],axis=1,inplace=True)
file_one_df.head()
```

```python
#store US population for 2010 to 2017
us_pop_df=file_one_df.iloc[[0,2]]
us_pop_df
```

```python
#rename columns
file_one_df = file_one_df.rename(columns={'GC.target-geo-id2':'Zip Code'})
file_one_df = file_one_df.rename(columns={'GC.display-label.1':'Metro Area'})
file_one_df = file_one_df.rename(columns={'respop72010':'Population 2010'})
file_one_df = file_one_df.rename(columns={'respop72011':'Population 2011'})
file_one_df = file_one_df.rename(columns={'respop72012':'Population 2012'})
file_one_df = file_one_df.rename(columns={'respop72013':'Population 2013'})
file_one_df = file_one_df.rename(columns={'respop72014':'Population 2014'})
file_one_df = file_one_df.rename(columns={'respop72015':'Population 2015'})
file_one_df = file_one_df.rename(columns={'respop72016':'Population 2016'})
file_one_df = file_one_df.rename(columns={'respop72017':'Population 2017'})
file_one_df.reset_index()

len(file_one_df)
file_one_df.head()
```

```python
file_one_df.drop([0,2], axis=0, inplace=True)
file_one_df.reset_index()
file_one_df.head()
```

```python
#for row 1 (US population), update zip to 00000
try:
    file_one_df['Zip Code'][1]  = '99999'
    #file_one_df.recet_index(inplace=True)
except:
    print('Cannot correct invalid zipcodes')
file_one_df.head()
```

```python
#clean up data: normilize Zip Codes
#drop rows where zip is NAAN 
try:
    file_one_df.dropna(subset=['Zip Code'], inplace=True) 
except:
    print(len(file_one_df))
file_one_df.head()
```

```python
#drop rows where zip is less then 5 digits
try:
    file_one_df.drop(file_one_df[file_one_df['Zip Code'].map(len) < 5].index, inplace=True)
except:
    print('Cannot remove invalid zipcodes')
print(len(file_one_df))
#file_one_df.head(50)
```

```python
#if zip is longer then 5 digits, read only the last 5 digits
try:
    file_one_df['Short Zip']  = file_one_df['Zip Code'].str[-5:]
except:
    print('Cannot correct invalid zipcodes')
file_one_df.head()
```

```python
#select towns where population was more then 1 mil in 2010
#len(file_one_df)  append#980
file_one_df[['Population 2010']] = file_one_df[['Population 2010']].apply(pd.to_numeric)
file_one_df[['Population 2011']] = file_one_df[['Population 2011']].apply(pd.to_numeric)
file_one_df[['Population 2012']] = file_one_df[['Population 2012']].apply(pd.to_numeric)
file_one_df[['Population 2013']] = file_one_df[['Population 2013']].apply(pd.to_numeric)
file_one_df[['Population 2014']] = file_one_df[['Population 2014']].apply(pd.to_numeric)
file_one_df[['Population 2015']] = file_one_df[['Population 2015']].apply(pd.to_numeric)
file_one_df[['Population 2016']] = file_one_df[['Population 2016']].apply(pd.to_numeric)
file_one_df[['Population 2017']] = file_one_df[['Population 2017']].apply(pd.to_numeric)

one_mil_df= file_one_df.loc[file_one_df['Population 2010'] > 1000000]
one_mil_df.reset_index(inplace=True)
one_mil_df        #75 cities >1 mil in 2010
```
